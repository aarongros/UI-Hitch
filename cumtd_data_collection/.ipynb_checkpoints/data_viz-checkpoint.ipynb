```r
# install.packages("RSQLite")
library("DBI")
```

```r
conn <- dbConnect(RSQLite::SQLite(), "stop_times.db")
res <- dbSendQuery(conn, "SELECT \"2018-11-10\" FROM stop_times WHERE \"2018-11-10\" IS NOT NULL")
vals <- dbFetch(res)
vals <- vals[['2018-11-10']]
summary(vals)
```

```r
hist(vals, breaks=1000, freq=FALSE, xlab="Bus Delay", main="Histogram of Bus Delays on 2018-11-10")
```

```r
hist(vals, breaks=1000, xlim=c(-60,60), ylim=c(0,0.05), freq=FALSE, xlab="Bus Delay", main="Histogram of Bus Delays on 2018-11-10")
```

```r
hist(vals, breaks=1000, ylim=c(0,0.005), freq=FALSE, xlab="Bus Delay", main="Histogram of Bus Delays on 2018-11-10")
```
